```python
with open('eng-kannada.txt','r',encoding='utf-8') as f:
    rows=f.read().split('\n')
```

```python
import pandas as pd
```

```python
# create a dataframe of all rows
df=pd.DataFrame(rows,columns=['text'])
```

```python
df['English']=df['text'].apply(lambda x: x.split('\t')[0])
df['Kannada']=df['text'].apply(lambda x: x.split('\t')[-1])
```

```python
df
```

```python
df.drop('text',axis=1,inplace=True)
```

```python
#save a csv file 
df.to_csv('eng-kannada.csv',index=False)
```

```python
#create a word cloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# create a word cloud of all words in the dataset
all_English_words=' '.join([text for text in df['English']])
wordcloud=WordCloud(width=800,height=500,random_state=21,max_font_size=110).generate(all_English_words)
```

```python
plt.figure(figsize=(10,7))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
all_Kannada_words=' '.join([text for text in df['Kannada']])
wordcloud=WordCloud(width=800,height=500,random_state=21,max_font_size=110).generate(all_Kannada_words)
plt.figure(figsize=(10,7))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```
